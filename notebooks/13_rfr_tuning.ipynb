```python
import sys
import os
from pathlib import Path
from dotenv import load_dotenv
import cudf
import optuna

sys.path.append(os.path.abspath(".."))

import src.models.rfr.rfr_optuna_optimizer as op
```

```python
# データの読み込み
env_path = Path.cwd().parent / ".env"
load_dotenv(dotenv_path=env_path)
url = os.environ.get("OPTUNA_STORAGE_URL")

tr_df8 = cudf.read_csv("../artifacts/features/tr_df8.csv").astype("float32")
```

```python
# tr_df8
objective = op.create_objective(tr_df8)

op.run_optuna_search(
    objective,
    n_trials=20,
    study_name="rfr_v2",
    storage=url,
    sampler=optuna.samplers.TPESamplers(seed=42)
)
```
