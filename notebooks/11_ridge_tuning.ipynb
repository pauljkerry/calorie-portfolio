```python
import sys
import os
from pathlib import Path
from dotenv import load_dotenv
import cudf
import optuna

sys.path.append(os.path.abspath(".."))

import src.models.ridge.ridge_optuna_optimizer as op
```

```python
# データの読み込み
env_path = Path.cwd().parent / ".env"
load_dotenv(dotenv_path=env_path)
url = os.environ.get("OPTUNA_STORAGE_URL")

tr_df7 = cudf.read_csv("../artifacts/features/tr_df7.csv")
```

```python
# tuning
objective = op.create_objective(tr_df7)

op.run_optuna_search(
    objective,
    n_trials=20,
    study_name="ridge_v2",
    storage=url,
    sampler=optuna.samplers.TPESampler(seed=42)
)
```
